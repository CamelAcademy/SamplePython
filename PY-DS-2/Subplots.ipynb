# Subplots


#### We'll explore another matplotlib-based stylistic feature: subplotting. Subplotting is a technique for creating multiple plots that live side-by-side in one overall figure. 

```python
import pandas as pd
reviews = pd.read_csv(r"C:\Users\User\Documents\wine\wine150k.csv", index_col = 0)
reviews.head(3)
```

```python
import matplotlib.pyplot as plt
fig, axarr = plt.subplots(2,1, figsize = (12,8)) #we got a figure with two rows and one column.
reviews['points'].value_counts().sort_index().plot.bar(ax = axarr[0])
reviews['province'].value_counts().head(20).plot.bar(ax = axarr[1])
```

```python
fig, axarr = plt.subplots(2,2, figsize = (12,8)) #we got a figure with two rows and one column.
reviews['points'].value_counts().sort_index().plot.bar(ax = axarr[0][0])
reviews['province'].value_counts().head(20).plot.bar(ax = axarr[1][1]) #the index is actually the position for the graphs
```

```python
fig, axarr = plt.subplots(2,2, figsize = (12,8)) #we got a figure with two rows and one column.

reviews['points'].value_counts().sort_index().plot.bar(ax = axarr[0][0], fontsize = 12, color = "mediumvioletred")
axarr[0][0].set_title("Wine Scores", fontsize = 18)


reviews['province'].value_counts().head(20).plot.bar(ax = axarr[1][1])
```

```python
fig, axarr = plt.subplots(2,2, figsize = (12,8)) #we got a figure with two rows and one column.

reviews['points'].value_counts().sort_index().plot.bar(ax = axarr[0][0], fontsize = 12, color = "mediumvioletred")
axarr[0][0].set_title("Wine Scores", fontsize = 18)

reviews['price'].value_counts().plot.hist(ax = axarr[0][1], fontsize = 12, color = "mediumvioletred")
axarr[0][1].set_title("Wine Price", fontsize = 18)

reviews['variety'].value_counts().head(20).plot.bar(ax = axarr[1][0], fontsize = 12, color = "mediumvioletred")
axarr[1][0].set_title("Varieties", fontsize = 18)

reviews['province'].value_counts().head(20).plot.bar(ax = axarr[1][1])
axarr[1][1].set_title("Wine Provinces", fontsize = 18)

plt.subplots_adjust(hspace = .3)

import seaborn as sns
sns.despine()
```

### Why are subplots useful?

* Oftentimes as a part of the exploratory data visualization process you will find yourself creating a large number of smaller charts probing one or a few specific aspects of the data. 

* Finally, subplots are critically useful because they enable faceting. Faceting is the act of breaking data variables up across multiple subplots, and combining those subplots into a single figure. So instead of one bar chart, we might have, say, four, arranged together in a grid.
