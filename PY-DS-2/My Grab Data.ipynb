### Splitting data into different data frames based on:
- DF1: Going back home 
- DF2: Going out from home 
- DF3: Going to school

Home = Vista Komanwel B,
School = APU New Campus

```python
df1 = george.drop(george[george['drop off'] != 'VISTA KOMANWEL B'].index) #back home
df2 = george.drop(george[george['pick up'] != 'VISTA KOMANWEL B'].index) #from home
df3 = george.drop(george[george['drop off'] != 'APU New Campus'].index) #to school
df1.head()
```

### Fare when going out from Home

```python
df = df2[df2['Fare'] > 0]
%matplotlib inline
fig = plt.figure(figsize=(20,15))
sns.boxplot(x = 'Drop Off', y = 'Fare', data = df)
sns.despine(offset=10, trim=True)
sns.despine(offset=10, trim=True)
plt.xticks(rotation=90)
```

```python
df1.Fare.describe()
```

### Going out from Home to destination

The data has been filtered out for GrabCar and GrabShare only

```python
df4 = df2[df2['Fare'] > 0] 
df5 = df4[df4['Booking Type'].isin(['GrabCar', 'GrabShare'])]
%matplotlib inline
fig = plt.figure(figsize=(20,15))
sns.boxplot(x = 'Drop Off', y = 'Fare', data = df5, hue = 'Booking Type')
sns.despine(offset=10, trim=True)
sns.despine(offset=10, trim=True)
plt.xticks(rotation=90)
```

### Going back to Home from Destination
Data has been filtered out for booking type in GrabCar or Grabshare only

```python
df4 = df1[df1['fare'] > 0] 
df5 = df4[df4['booking type'].isin(['GrabCar', 'GrabShare'])]
%matplotlib inline
fig = plt.figure(figsize=(20,15))
sns.boxplot(x = 'pick up', y = 'fare', data = df5, hue = 'booking type')
sns.despine(offset=10, trim=True)
sns.despine(offset=10, trim=True)
plt.xticks(rotation=90)
```

### Fare against Ride Distance based on Booking Type

```python
sns.lmplot(x = 'Ride Distance (km)', y = 'Fare', hue = "Booking Type",markers = ['x','o'], data = df5) 
```
