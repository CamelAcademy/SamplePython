```python
footballers = df.copy()
footballers['Unit'] = df['Value'].str[-1]
footballers['Value (M)'] = np.where(footballers['Unit'] == '0', 0, 
                                   footballers['Value'].str[1:-1].replace(r'[a-zA-Z]',''))
footballers['Value (M)'] = footballers['Value (M)'].astype(float)
footballers['Value  (M)'] = np.where(footballers['Unit'] == 'M',
                                    footballers['Value (M)'],
                                    footballers['Value (M)']/1000)
footballers = footballers.assign(Value =footballers['Value (M)'],
                                Position = footballers['Preferred Positions'].str.split().str[0])
```

## Shows the top 4 players

This are multivariate data thus a visual variable is used. Visual variable is any visual dimentsion or marker that can be use to perceptually distinguish between two data elements from one another. 

# 1. Multivariate Scatter Plots

To see which offensive player gets paid the most: Striker, right winger or left winger.

```python
import seaborn as sns

sns.lmplot(x='Value', y='Overall', hue='Position', 
           data=footballers.loc[footballers['Position'].isin(['ST', 'RW', 'LW'])], 
           fit_reg=False)
```

The scatter plots shows three variables. Value, overall and color, whereby color is the "hue" parameter that tracks the three categories of interest the player the point represents in. 

A new variable is the chart is color, eventho its tricky to use. Overplotting do occur. 

More than that, we can introduce a shape variable as well.

### Using Shape as a variable

```python
sns.lmplot(x='Value', y='Overall', markers=['o', 'x', '*'], hue='Position',
           data=footballers.loc[footballers['Position'].isin(['ST', 'RW', 'LW'])],
           fit_reg=False
          )
```

# 2. Grouped Box plot
A plot that takes advantage of grouping. Suppose we have a question: Do strikers score higher on aggression than goalkeepers do?

```python
f = (footballers.loc[footballers['Position'].isin(['ST','GK'])].loc[:,['Value','Overall','Aggression','Position']])

f = f[f["Overall"] >= 80]
f = f[f["Overall"] < 85]
f['Aggression'] = f['Aggression'].astype(float)

sns.boxplot(x = 'Overall', y='Aggression', hue = 'Position', data = f)
```

The plots has shown that indeed, goalkeepers has a lower aggression scores than strikers. 
Grouping is extremely communicative visual variable, but it has low cardinality. Hard to group more than a handful of categorical value.

# 3. Summarization 

It is hard to squeeze enough dimensions onto a plot without sacrificing interpretability. Thus busy plots are hard to interpret. Summarization is the creation and addition of new variables by mixing and matching the information provided in the old ones.


## 3.1 Heatmap
Most heavily used summarization visualization is the correlation plot, measuring the correlation between every pair of values in a dataset and plot result in color.

```python
f = (
    footballers.loc[:, ['Acceleration', 'Aggression', 'Agility', 'Balance', 'Ball control']]
        .applymap(lambda v: int(v) if str.isdecimal(v) else np.nan)
        .dropna()
).corr()

sns.heatmap(f, annot=True)
```

Each plot is intersection of two variables, the color and label indicate the amount of correlation between the two variables. For example, agility and acceleration is highly correlated. 

## 3.2 Parallel Coordinates
A parallel coordinates plot provides another way of visualizing data across many variables


```python
from pandas.plotting import parallel_coordinates

f = (footballers.iloc[:,12:17].loc[footballers['Position'].isin(['ST','GK'])].applymap(lambda v: int(v) if str.isdecimal(v) else np.nan).dropna()
    )
f['Position'] = footballers['Position']
f = f.sample(200)

parallel_coordinates(f,'Position')
```

Showing 200 goalkeepers in dark green and Strikers in Light green across 5 variables.

Parallel coordinates are great for determining how distinguishable different class are in data. The standardize variables from top to bottom! 

As we can see, strikers are almost uniformly higher ratted on all variables, easily distinguishing two classes.
