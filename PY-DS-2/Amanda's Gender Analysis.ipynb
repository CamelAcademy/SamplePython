## Plotting scatter plots with salary against age based on gender

```python
sns.lmplot(x='age', y='salary', hue='gender', 
           data=gender.loc[gender['gender'].isin(['Male', 'Female'])], 
           fit_reg=False)
```

## A Distribution plot of salary and age, with kde true

KDE, short for "kernel density estimate", is a statistical technique for smoothing out data noise. It addresses an important fundamental weakness of a line chart: it will buff out outlier or "in-betweener" values which would cause a line chart to suddenly dip.
A KDE plot is better than a line chart for getting the "true shape" of interval data.

```python
sns.distplot(gender['salary'], kde = True)
```

```python
sns.distplot(gender['age'], kde = True)
```

## Kde true below as well

```python
sns.kdeplot(gender.salary)
```

## Salary mean below across all gender

```python
gender['salary'].mean()
```

## A hex plot of salary and age 

Surprisingly, large amount of workers regardless of gender, are concentrated at age of 26. 

```python
sns.jointplot(x = 'age', y = 'salary', data = gender, kind = 'hex', gridsize = 10)
```

## Boxplots for salary and gender

Male has higher average salary compared to female, without skill level being put into consideration. Does this means that it is the descrimination has occured?

```python
sns.boxplot(x= 'gender', y = 'salary', data = gender)
```

## Using Bivariate KDE chart, Salary and Age


Bivariate KDE plots like this one are a great alternative to scatter plots and hex plots. They solve the same data overplotting issue that scatter plots suffer from and hex plots address, in a different but similarly visually appealing.

```python
sns.kdeplot(gender[['age', 'salary']])
```

## Boxplot for salary against skill based on gender

Skills are based on three levels, with 3 being the highest skilled jobs

```python
sns.boxplot(x = "skill", y = "salary", hue = 'gender', data = gender)
```
