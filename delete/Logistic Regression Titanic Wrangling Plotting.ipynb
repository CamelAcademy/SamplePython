```python
df1=pd.get_dummies(df0, columns=['Pclass','Sex', 'Embarked'], drop_first=True)
df1
```

```python
df1.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis='columns', inplace=True)
df1.dropna(inplace=True)
```

```python
df1.info()
```

```python
df1.isnull().sum()
```

## Training and Testing with Data

```python
X=df1.drop('Survived',axis='columns')
y=df1.Survived
```

```python
from sklearn.model_selection import train_test_split
X_learn, X_actual, y_learn, y_actual = train_test_split(X, y, test_size=0.25)
```

```python
X_learn
```

```python
from sklearn.linear_model import LogisticRegression
lrModel = LogisticRegression()
lrModel.fit(X_learn, y_learn)
```

```python
print('Y-Intercept: '+ str(lrModel.intercept_) + '\n\nX-Cofficient: \n' + str(lrModel.coef_))
```

```python
y_predict = lrModel.predict(X_actual)
y_predict
```

```python
# Feature Variable - Independent - X variable
# Target Variable  - Dependent   - Y variable
from sklearn.metrics import mean_squared_error, accuracy_score 

print ('Mean Square Error - Actual Y & Predicted Y ........ : %8.8f' % mean_squared_error(y_actual, y_predict))
print ('Model Score - Actual X & Actual Y ................. : %8.8f' % lrModel.score(X_actual, y_actual))
print ('Accuracy Score - Actual Y & Predicted Y ........... : %8.8f' % accuracy_score(y_actual, y_predict))
```
