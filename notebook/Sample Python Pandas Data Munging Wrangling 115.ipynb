```python
import pandas as pd
```

```python
#df from a csv file, 1st row would be header with no skips or default header is zero
df1 = pd.read_csv('countries_short.csv')
df1
```

```python
#df from a csv file, 3rd row would be the header after skipping 2 rows including the header
df2 = pd.read_csv('countries_short.csv', skiprows=2) 
df2.head()
```

```python
#df from a csv file, 3rd row in the file would be the header
df3 = pd.read_csv('countries_short.csv', header=1)
df3.head()
```

```python
#df from a csv file, skip 2 rows in the file and then 2nd row would be the header
df4 = pd.read_csv('countries_short.csv', skiprows=2, header=1) 
df4.head()
```

```python
#After skipping 1 row we lost the header, then we can pass with header=None and column names as argument
df5 = pd.read_csv('countries_short.csv', skiprows=1, header=None, names=["NAME","YEAR","POPULATION"])
df5.head()
```

```python
#If you need to read only n rows with header then read with nrows
df6 = pd.read_csv('countries_short.csv', nrows=3) 
df6.head()
```

```python
#If you like to replace all invalid numbers to NaN
df7 = pd.read_csv('countries_short.csv', na_values=["NA","N/A", "Not Available", "?"]) 
df7
```

```python
# Munging/Wrangling We can also do different column level validation with a dictionary
df8 = pd.read_csv('countries_short.csv', na_values =
                    {  'country': ["NA","N/A", "Not Available", "?", "Unknown"],
                        'year': ["NA","N/A", "Not Available", "?", 0],
                        'population': ["NA","N/A", "Not Available", "?"],
                    }
                 )
df8
```
