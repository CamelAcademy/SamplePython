```python
import pandas as pd
import random
 
# Read CSV would read the CSV file over the internet too.
df0 = pd.read_csv('https://s3-eu-west-1.amazonaws.com/shanebucket/downloads/uk-500.csv')
rows, columns = df0.shape
print('Total # of rows: {} and columns: {} in DataFrame df0'.format(rows, df0.shape[1]))

# set a numeric id for use as an index for examples.
df0['rand'] = [random.randint(10,99) for x in range(df0.shape[0])]
 
df0
```

```python
# Single Row selection from DataFrame using iloc. Note output data type is Series.
df0.iloc[0] # first row of data frame (Aleshia Tomkiewicz)
```

```python
# Single Row selection from DataFrame using iloc. Note output data type is Series.
df0.iloc[-1] # last row of DataFrame (Mi Richan)
```

```python
# Single Row selection from DataFrame using iloc. Note output data type is Series.
df0.iloc[-2] # second last row of DataFrame (Celestina Keeny)
```

```python
# Single Column selection from DataFrame using iloc. Note output data type is Series.
# first column first_name of DataFrame
df0.iloc[:,0]
```

```python
# Single Column selection from DataFrame using iloc. Note output data type is Series.
# last column rand of DataFrame
df0.iloc[:,-1]
```

```python
# Double Column selection from DataFrame using iloc. Note output data type is DataFrame.
# first & second column first_name & last_name of DataFrame
df0.iloc[:,[0,1]]
```

```python
# Double Column selection from DataFrame using iloc. Note output data type is DataFrame.
# first & second column but not the third column of DataFrame
df0.iloc[:, 0:2]
```

```python
# Selection of multiple rows and all columns with iloc. Note output data type is DataFrame.
df0.iloc[0:5]
```

```python
# Selection of multiple rows (1st, 6th, 11th & 16th) and columns(1st, 2nd & 5th) with iloc
df0.iloc[[0,5,10,15], [0,1,4]] # Note output data type is DataFrame.
```

```python
# Selection of multiple rows(1st to 5th but not 6th) and columns(1st to 3rd but not 4th) with iloc
df0.iloc[0:5, 0:3] # rows(starting from 0, 5 of them) columns (starting from 0, 3 of them)
```

```python
# Select rows with index values Aleshia, Evan and Ulysses, with all columns between 'city' and 'web'
df0.loc[['Aleshia', 'Ulysses','Evan'], 'city':'web']
```

```python
# Select rows, with first name in between 2 and columns company_name, city and web.
df0.loc['Aleshia':'Ulysses', ['company_name', 'city', 'web']]
```

```python
# Select rows & columns with between 2 values.
df0.loc['Aleshia':'France', 'city':'web']
```

```python
df0.loc[['Tyisha','Charlette','Mi'],['last_name','county']]
```

```python
# Boolean / Logical indexing using loc
df0.loc[df0['last_name']=='Veness', ['company_name','county']]
```

```python
# Returns a Series
df0.loc[df0['city']=='Abbey Ward', 'county']
```

```python
# Returns a DataFrame as input is a list
df0.loc[df0['city']=='Abbey Ward', ['county']]
```

```python
# Select rows with email id ulysses@hotmail.com and all columns between 'email' and 'city'
df0.loc[df0['email'] == 'ulysses@hotmail.com', 'email':'city']
```

```python
# Select rows where the email column ends with 'hotmail.com' and include all columns
df0.loc[df0['email'].str.endswith("hotmail.com")]
```

```python
# Select rows with last_name equal to isin values
df0.loc[df0['last_name'].isin(['Veness', 'Richan', 'Eric'])]  
```

```python
# Select rows with last name Veness AND email id ends with hotmail.com
df0.loc[df0['email'].str.endswith("hotmail.com") & (df0['last_name'] == 'Veness')]
```

```python
# select rows with ununique rand column between 100 and 200, and just return 'postal' and 'web' columns
df0.loc[(df0['rand'] >= 24) & (df0['rand'] <= 25), ['postal', 'web','rand']] 
```

```python
# Inner lambda function will returns True/False value for each row.
# Then select the rows where the company name is made of 4 words.
df0.loc[df0['company_name'].apply(lambda x: len(x.split(' ')) == 4)]
```
