```python
#df from read csv. Parse the date to change Scring to Date format
df1 = pd.read_csv('location_weather_short.csv', parse_dates=['Date'])
df1
```

```python
print(type(df1))
print(type(df1.Date))
print(type(df1.Date[0]))
print(type(df1.Location))
print(type(df1.Location[0]))

df1.dtypes
```

```python
df1.set_index("Date", inplace=True)
df1
```

```python
df1.dtypes
```

```python
#Replace all NaN with Zero
df2 = df1.fillna(0)
df2
```

```python
# Replace different values for different columns
df3 = df1.fillna({  'Location':'Unknown Location',
                    'WindSpeed': df1.WindSpeed.mean(), 
                    'Precipitation': df1.Precipitation.mean(), 
                    'tMax': df1.tMax.mean(),
                    'tMin': df1.tMin.mean() })
df3
```

```python
# forword fill by carry forward values across rows/days
df4=df1.fillna(method="ffill") # axis="rows"
df4
```

```python
# backward fill going across columns instead of rows
df5=df1.fillna(method="bfill", axis="columns")
df5
```

```python
# forward fill across rows with limit of 1 row; don't assume more than 1 row :)
df6=df1.fillna(method="ffill", limit=1)
df6
```

```python
# Simple and linear Interpolation to guess 1 or more numbers. Also checkout the numbers in last rows
df7=df1.interpolate()
df7
```

```python
#drop rows if any one cell is N/A
df8 = df1.dropna()
df8
```

```python
#drop very bad data points. drop only if all non key/index fields are NA
df9 = df1.dropna(how='all')
df9
```

```python
df9 = df1.dropna(thresh=3) # We need 3 or more valid values for each row
df9
```

```python
dt = pd.date_range("2019-12-09","2019-12-21")
idx = pd.DatetimeIndex(dt)
df10 = df1.reindex(idx)
df10
```

```python
import numpy as np
df11 = df10.replace(0.26, np.NaN) # search and replace all 0.26 with NaN
df11
```

```python
df12 = df10.replace([0.26,35], np.NaN) # Search and replace 0.26 and 35 with NaN
df12
```

```python
df13 = df10.replace({'Location':np.NaN}, 'NYC') # Replace Location column NaN with NYC
df14 = df13.replace({'WindSpeed': 6.040,'Precipitation': 1.570}, np.NaN) #Replace multiple columns with NaN
df14
```

```python
# better guess based on time interpolation for all number columns.
df15=df14.interpolate(method="time")
df15
```

```python
# Finally replace different values for String and number cells :)
df16 = df15.replace({'NYC':'NEW YORK CITY', np.NaN: 99.99})
df16
```
