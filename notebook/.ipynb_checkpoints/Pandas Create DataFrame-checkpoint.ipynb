```python
import pandas as pd
# https://pandas.pydata.org/pandas-docs/stable/user_guide/io.html
```

```python
#df from a csv file
df0 = pd.read_csv("countries_year_population.csv")
df0.head()
```

```python
#df from a excel file
df1 = pd.read_excel("countries_year_population.xlsx","cyp")
df1.tail()
```

```python
#df from records - list of tuples for each row.
data = [
    ('Afghanistan', 1997, 22227415),
    ('Australia', 2002, 19546792),
    ('Bangladesh', 1952, 46886859),
    ('Canada', 2007, 33390141),
    ('China', 1952, 556263527),
    ('Malaysia', 1977, 12845381),
    ('Pakistan', 1972, 69325921),
    ('United States', 1987, 242803533)]

# we need to provide column name as argument as it's not in the tuple/data
df2 = pd.DataFrame(data, columns=["country","year","population"])
df2
```

```python
#df using dictionary{key,values} - using python dictionary for each column/Series
data = {
    'country':['Afghanistan', 'Australia', 'Belgium', 'Brazil', 'Egypt', 'Ethiopia', 'Cambodia', 'Zimbabwe'],
    'year':[2007, 1977, 1972, 1962, 1962, 2007, 1957, 1962],
    'population':[31889923, 14074100, 9709100, 76039390, 28173309, 76511887, 5322536, 4277736]}

df3 = pd.DataFrame(data)
df3
```

```python
#df from dictionary{key,values} - using python dictionary for each column/Series
data = {
    'country':['Afghanistan', 'Australia', 'Belgium', 'Brazil', 'Egypt', 'Ethiopia', 'Cambodia', 'Zimbabwe'],
    'year':[2007, 1977, 1972, 1962, 1962, 2007, 1957, 1962],
    'population':[31889923, 14074100, 9709100, 76039390, 28173309, 76511887, 5322536, 4277736]}

df4 = pd.DataFrame.from_dict(data)
df4
```

```python
#df from dictionary{key,values} with orient='index' - using python dictionary for each row
data = {
    'Row-1':['Afghanistan', 1997, 22227415],
    'Row-2':['Australia', 2002, 19546792],
    'Row-3':['Bangladesh', 1952, 46886859],
    'Row-4':['Canada', 2007, 33390141],
    'Row-5':['China', 1952, 556263527],
    'Row-6':['Malaysia', 1977, 12845381],
    'Row-7':['Pakistan', 1972, 69325921],
    'Row-8':['United States', 1987, 242803533]}

df5 = pd.DataFrame.from_dict(data, orient='index')
df5
```

```python
#df from dictionary{key,values} with columns name argument - using python dictionary for each row
data = {
    'Row-1':['Afghanistan', 1997, 22227415],
    'Row-2':['Australia', 2002, 19546792],
    'Row-3':['Bangladesh', 1952, 46886859],
    'Row-4':['Canada', 2007, 33390141],
    'Row-5':['China', 1952, 556263527],
    'Row-6':['Malaysia', 1977, 12845381],
    'Row-7':['Pakistan', 1972, 69325921],
    'Row-8':['United States', 1987, 242803533]}

df6 = pd.DataFrame.from_dict(data, orient='index', columns=["country","year","population"])
df6
```

```python
#df from list of multiple dictionary. You can skip or change order of dictionary :)
data = [
    {'country':'Afghanistan', 'year':1997, 'population':22227415},
    {'year':2002, 'population':19546792},
    {'country':'Bangladesh', 'population':46886859},
    {'country':'Canada', 'year':2007},
    {'country':'China', 'year':1952, 'population':556263527},
    {'country':'Malaysia', 'year':1977, 'population':12845381},
    { 'year':1972, 'population':69325921,'country':'Pakistan'},
    {'country':'United States', 'year':1987, 'population':242803533}]

df7 = pd.DataFrame(data)
df7
```
