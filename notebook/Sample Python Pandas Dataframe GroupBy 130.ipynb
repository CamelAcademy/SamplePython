```python
import pandas as pd
```

```python
#df from read csv file with Index Data Values and Volumes
df0 = pd.read_csv('yahoo-futures.csv')
df0.head()
```

```python
# For each date find average and then print top 5 rows
dfg1 = df0.groupby('Date')
dfg1.mean().head()
```

```python
# To print (Key) Date and (Value) DataFrame grouped by Date
for Date, dfDate in dfg1:
    print(Date) # String DataType
    print(dfDate) # DataFrame DataType dfFutures.dtypes
```

```python
dfg1.get_group('3/16/2020') # to access DataFrame for a Date, Key-Value pair
```

```python
dfg2 = df0.groupby('Futures')
for Futures, dfFutures in dfg2:
    print(Futures)
    print(dfFutures)
```

```python
# Splitting the data into groups based on some criteria.
# Applying a function to each group independently.
# Combining the results into a data structure.

# Watchout - Cell in each row are from different dates, as 
# its not required market should open & close low on same date

dfMin = dfg2.min()
dfMin
```
